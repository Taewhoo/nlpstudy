뿌리 단어를 찾아 다른 형태의 단어들을 합치는 것

ex. am, are, is -> be (표제어)

표제어 추출 위해서는 형태학적 분석(parsing) 먼저 진행해야.

어간(stem): 단어 의미 담는 핵심 부분 (cat)

접사(affix): 단어 추가적 의미 부분 (s)


어간 추출과는 달리, 표제어 추출은 해당 단어의 품사 정보를 보존한다. (POS 태그 보존)


```python
import nltk
nltk.download('wordnet') #표제어 추출 도구
```

(6) 정수 인코딩: 각 단어를 고유한 정수에 맵핑시킨다.

컴퓨터는 텍스트보다 숫자를 더 잘 처리한다.

```python
import nltk
nltk.download('punkt')
```

```python
# dictionary 사용
from nltk.tokenize import word_tokenize
from nltk.tokenize import sent_tokenize
from nltk.corpus import stopwords
```

```python
import nltk
nltk.download('stopwords')
```

```python
raw_text = """A barber is a person. a barber is good person.
a barber is huge person. he Knew A Secret!
The Secret He Kept is huge secret. Huge secret.
His barber kept his word. a barber kept his word.
His barber kept his secret. But keeping and keeping
such a huge secret to himself was driving the barber crazy.
the barber went up a huge mountain."""
```

```python
sentences = sent_tokenize(raw_text)
print(sentences)
```

```python
vocab = {}
preprocessed_sentences = []
stop_words = set(stopwords.words('english'))

for sent in sentences: #각문장
  word = word_tokenize(sent) #각문장 내 단어들

  result = []
  for w in word: #각 단어
    w = w.lower() #단어 소문자화
    if w not in stop_words: #불용어 제거
      if len(w) > 2: # 단어 길이 제어
        result.append(w)
        if w not in vocab:
          vocab[w] = 0
        vocab[w] += 1        # 단어별 빈도수
  preprocessed_sentences.append(result)
print(preprocessed_sentences)
```

```python
print(vocab)
```

```python
vocab_sort = sorted(vocab.items(), key = lambda x:x[1], reverse = True)
print(vocab_sort)
# counter를 사용하는 것이 더욱 편리하다.
```

```python
# 빈도수 순위로 다르게 표현
word_rank = {}
i = 0
for (word, frequency) in vocab_sort:
  if frequency > 1: # 빈도수 적은 단어 제외
    i = i + 1
    word_rank[word] = i # {word, i} 구조 확립

print(word_rank)
```

```python
# 상위 5개만 추출
vocab_size = 5
low_ranked_words = [word for word, rank in word_rank.items() if rank > vocab_size]
for w in low_ranked_words:
  del word_rank[w]  #word_rank dictionary에서 w에 해당하는 인덱스 정보 삭제
print(word_rank)
```

```python
preprocessed_sentences
# 2번째 문장의 'good'은 word_rank에 존재하지 않는다. -> OOV (out-of-vocabulary)
```

```python
word_rank['OOV'] = len(word_rank) + 1
print(word_rank)
```

```python
#정수 인코딩
encoded_sentences = []
for wordlist in preprocessed_sentences:
  encoded_sentence = []
  for word in wordlist:
    try:
      encoded_sentence.append(word_rank[word]) #word rank 단어의 index 추가
    except KeyError:
      encoded_sentence.append(word_rank['OOV'])
  encoded_sentences.append(encoded_sentence)
print(encoded_sentences)

```

원-핫 인코딩

단어 집합: 중복되지 않는 단어들의 모음

단어 집합의 크기 = 벡터로 표현.

표현하고 싶은 단어의 인덱스 = 1, 나머지는 0

이렇게 표현된 벡터를 원-핫 벡터라고 한다.


---


(1) 고유 인덱스 부여 (정수 인코딩)

(2) 표현하고 싶은 단어의 인덱스 = 1, 다른 단어에는 0 부여

```python
# example

#토큰화
from konlpy.tag import Okt
okt = Okt()
token = okt.morphs("나는 자연어 처리를 배운다")
print(token)
```

```python
word2index = {}
for voca in token:
  if voca not in word2index.keys():
    word2index[voca] = len(word2index) #0부터 시작해서 인덱스 배정
print(word2index)
```

```python
def one_hot_encoding(word, word2index):
  one_hot_vector = [0]*(len(word2index)) # [0,0,0,0,0,0]
  index = word2index[word] #인덱스 = 문장 내 단어 순서
  one_hot_vector[index] = 1 # [0,0,1,0,0,0]
  return one_hot_vector

one_hot_encoding('자연어', word2index)
```

```python
import urllib.request
from soynlp import DoublespaceLineCorpus
from soynlp.word import WordExtractor

#학습 기반이므로 데이터 다운로드
urllib.request.urlretrieve("https://raw.githubusercontent.com/lovit/soynlp/master/tutorials/2016-10-20.txt", filename="2016-10-20.txt")
```

```python
corpus = DoublespaceLineCorpus('2016-10-20.txt') # 다수의 문서로 데이터 분리
len(corpus)
```

```python
# 3개 문서만 출력
i = 0
for document in corpus:
  if len(document) > 0:
    print(document)
    i = i + 1
  if i == 3:
    break
```

```python
word_extractor = WordExtractor()
word_extractor.train(corpus)
score_table = word_extractor.extract()
```

soynlp의 응집확률: 높을수록 그 문자열이 하나의 단어로 등장할 가능성이 높아진다.

soynlp - branching entropy

```python
# 문자열에서 다음 문자가 등장할 수 있는 정도를 나타낸다.
# ex. 디(?) -> 디스(..?) -> 디스플(아~) -> 디스플레이
```

```python
score_table['디스'].right_branching_entropy
# '디스' 뒤에는 다양한 문자가 올 수 있다.
```

```python
score_table['디스플'].right_branching_entropy
# '디스플' 이후에는 '레'가 오는 것이 너무나도 명백하므로 0 반환.
```

```python
score_table['디스플레이'].right_branching_entropy
# 단어가 끝나면 조사 등이 붙을 수 있으므로 값이 증가한다.
```

LTokenizer: 점수가 가장 높은 L토큰(응집도가 큰, 즉 단어)을 기준으로 분리한다.

```python
from soynlp.tokenizer import LTokenizer
scores = {word: score.cohesion_forward for word, score in score_table.items()}
# score.cohesion_forward : 응집도
l_tokenizer = LTokenizer(scores = scores)
l_tokenizer.tokenize('국제사회와 우리의 노력들로 범죄를 척결하자', flatten = False)
```

```python
# 최대 점수 tokenizer : 띄어쓰기 안된 문장에서, 점수가 높은대로 순차적으로 찾아냄.
from soynlp.tokenizer import MaxScoreTokenizer

maxscore_tokenizer = MaxScoreTokenizer(scores = scores)
maxscore_tokenizer.tokenize('국제사회와우리의노력들로범죄를척결하자')
```
