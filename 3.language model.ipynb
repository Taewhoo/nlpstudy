**언어 모델링(-ing)** : 주어진 단어들로 아직 모르는 단어를 예측하는 작업.



# 2. 확률의 역할

a. 기계 번역

* P(나는 버스를 탔다) > P(나는 버스를 태운다)

b. 오타 교정

* 선생님이 교실로 부리나케

* P(달려갔다) > P(잘려갔다)

**c. 음성 인식**

* P(나는 메롱을 먹는다) < P(나는 메론을 먹는다)


# 3. 다음 단어 예측하기: 가장 보편적

a. 단어 시퀀스의 확률 (W:문장, w:개별단어)

* P(W) = P(w1,w2,.....,wn)

b. 다음 단어 등장 확률

**P(wn|w1,w2,....,wn-1)**: 이전 단어들이 나열된 상태에서, n번째 단어가 올 확률

즉 P(W) =  Σ i=1~n P(wi|w1,w2,....,w(n-1)

*2) 통계적 언어 모델*

# 1. 조건부 확률

* P(B|A) = P(A,B)/P(A)

* P(A,B) = P(A)P(B|A)

* P(x1,x2,.....,xn) = P(x1)P(x2|x1)P(x3|x1,x2).....P(xn|x1,x2,...,x(n-1))

# 2. 문장에 대한 확률

P(An adorable little boy is spreading smiles) = P(An)P(adorable|An)P(little|An adorable)...P(smiles|An adorable little boy is spreading)

즉 문장의 확률을 위해서는 **각 단어의 예측 확률들을 곱한다.**

# 3. 카운트 기반의 접근

그럼 각 단어의 예측 확률들은 어떻게 알까?

P(is|An adorable little boy) = count(An adorable little boy + is) / count(An adorable little boy)

ex. an adorable little boy- 100번

an adorable little boy is - 30번

**-> P(is|An adorable little boy) - 30%**


# 4. 한계점

- count 방식: 정말 많은 데이터 필요

- 희소 문제: 훈련 코퍼스에 'an adorable little boy(is)' 라는 시퀀스가 없다면, 확률은 0이 되거나 정의되지 않는다.

통계 모델 -> 인공 신경망 언어 모델으로 전환되는 이유

3) *N-gram 언어 모델*: 모든 단어가 아닌, 일부 단어(n개)만 고려하는 방법

# 1. 희소 문제 감소

P(is|An adorable little boy) -> P(is|boy)

# 2. N-gram

코퍼스에서 n개의 단어 뭉치 단위로 끊어, 이를 하나의 토큰으로 간주한다.

unigrams: an, adorable, little, ...

bigrams: an adorable, adorable little, little boy, ...

trigrams: an adorable little, adorable little boy, ...

4-grams: ~

ex. what comes after 'spreading'?

-> P(w|boy is spreading) = count(boy is spreading w) / count(boy is spreading)

# 3. N-gram의 한계점

* 희소 문제: SLM에 비해 낫지만, 여전히 존재.

* trade-off : n이 커질수록 정확, but 희소 문제는 커진다. (5 넘게 안하는걸 권장)


# 4. 적용 분야에 맞게 코퍼스를 수집하면, 당연히 정확한 언어 모델 생성 가능.

때로는 약점으로 지적되기도 한다. (훈련용 코퍼스에 따라 성능이 달라지므로)

*4)한국어 언어 모델: 영어보다 훨씬 어렵다.*

1. 어순이 중요하지 않다. (어떤 단어도 나타날 수 있다.)

2. 한국어는 교착어이다. (그녀가,그녀의,그녀로,그녀처럼,....) (=조사가 붙으므로 분리하는 토큰화 필요)

3. 띄어쓰기가 잘 지켜지지 않는다.

5) *Perplexity* (PPL): 헷갈리는 정도


모델의 성능을 일일이 비교하는건 힘들다. 

대신 내부적으로 성능을 수치화하여 결과를 내는 perplexity를 사용한다.


1. PPL은 낮을수록 성능이 좋음을 의미한다.

2. PPL은 분기 계수(가능한 경우의 수를 의미)이다.

ex. PPL = 10: 다음 단어를 예측하는 모든 시점마다 평균적으로 10개의 단어로 정답을 고민한다.

(PPL 10 -> 확률은 1/10)
